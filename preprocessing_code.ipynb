```python
import pandas as pd
#import file
reactions = pd.read_csv('/Users/candicewu/Desktop/DA Projects/6./Reactions.csv')
reactions.head()
```

```python
#delete the first column by using ilo[]
reactions=reactions.iloc[:,1:]
reactions.head()
```

```python
#inspect data type for each column and the number of missing values
reactions.info(verbose=True, show_counts=True)
```

According to the output, the total number of rows should be 25553. However, there are only 24573 rows for 'Type' column, which suggests that there may be some missing values in this column

```python
#remove missing values by using the dropna() function
reactions=reactions.dropna(subset=['Type'])
reactions.info()
```

```python
#delect irrelevant columns by using the drop() fcunction
reactions=reactions.drop(columns=['User ID'])
reactions.info()
```

```python
#rename column name by using the rename() function
reactions=reactions.rename(columns={"Type": "Reaction Type"})
```

```python
#import file
content = pd.read_csv('/Users/candicewu/Desktop/DA Projects/6./Content.csv')
content.head()
```

```python
#delete the first column by using ilo[]
content=content.iloc[:,1:]
content.head()
```

```python
content.info()
```

```python
#remove irrelevant columns by using the drop() function
content=content.drop(columns=['URL', 'User ID'])
content.info()
```

```python
#replaces a specified phrase with another specified phrase by using the replace() function
content['Category']=(content['Category'].str.replace(" "" ", "",regex=True))
content
```

```python
#rename column names by using the rename( ) function
content= content.rename(columns={"Type": "Content Type"})
content.head()
```

```python
#import file
reactiontypes = pd.read_csv('/Users/candicewu/Desktop/DA Projects/6./ReactionTypes.csv')
reactiontypes.head()
```

```python
#delection the first column by using the iloc() function
reactiontypes=reactiontypes.iloc[:,1:]
reactiontypes.head()
```

```python
#check
reactiontypes.info()
```

```python
#to merge two dataset by using the merge () function
df = pd.merge(
    reactions,
    reactiontypes,
    left_on=["Reaction Type"],
    right_on=["Type"],
    how="left",
)
df
```

```python
df.info()
```

```python
final_df = pd.merge(
   df,
   content,
    on=["Content ID"],
    how="left",
)
final_df
```

```python
final_df.info()
```

```python
final_df.to_csv("/Users/candicewu/Desktop/cleaned.csv", index=False)
```
